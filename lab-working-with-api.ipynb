#### Some cool functions ~to make things easier~

```python
# headers won't change
headers = {
    'x-rapidapi-host': "aerodatabox.p.rapidapi.com",
    'x-rapidapi-key': "9613ca6ca5msh62659872159916fp1a0a2cjsndd9a3e6ff324"
    }


def get_codes():

    location = input('Insert location:') #gets location name 
    print()

    url = "https://aerodatabox.p.rapidapi.com/airports/search/term"
    querystring = {"q":location.lower(), "limit":"10"} #searches airports by location from input

    response = requests.get(url, headers=headers, params=querystring) # getting responce of our querry
    codes = response.json()['items'] if 'items' in response.json() else location 
    # adds items from json file of response to codes variable *items include codes, name, shortname etc* else calls location input again
    return codes


def choose_code():

    codes = get_codes() # calls previous finction to find out codes based on location

    while type(codes) is str or len(set([code['shortName'] for code in codes])) != 1:
        #while gets more than 1 item in codes back or some text as a result

        if not codes:
            print('Try another location. Unable to find code for:', codes, '\n') 
            # if gets null value instead - outputs text and runs codes again
        else:
            print('More than one location found. Insert one of the following airports:', [code['shortName'] for code in codes], '\n')
            #error handling for more than one location in items
        codes = get_codes()

    return codes[0]['icao'] #returns icao code


def flight_distance_time(from_, to_):
    url = f"https://aerodatabox.p.rapidapi.com/airports/icao/{from_}/distance-time/{to_}" # access via end point change
    #instead of passing parameters in response
    response = requests.get(url, headers=headers)
    return pd.json_normalize(response.json()).T
```

```python
print('[DEPARTURE]')
from_ = choose_code()
print('[DESTINATION]')
to_ = choose_code()

flight_df = flight_distance_time(from_, to_)
flight_df
```
